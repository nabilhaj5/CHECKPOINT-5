```python
# PRUEBA 1: Llueve
esta_lluviendo = True
if esta_lluviendo:
    print("🟢 SÍ - Coger paraguas")
else:
    print("🟡 NO - No coger paraguas")
# Resultado: 🟢 SÍ - Coger paraguas
```

```python
# PRUEBA 2: No llueve
esta_lluviendo = False
if esta_lluviendo:
    print("🟢 SÍ - Coger paraguas")
else:
    print("🟡 NO - No coger paraguas")
# Resultado: 🟡 NO - No coger paraguas
```
