```python
#pregunta-2
#Cree una función de Python llamada suma que tome 3 argumentos 
# y devuelva la suma de los 3.

def suma(a,b,c):
    return a+b+c
suma(4,5,6)
```

```python
#pregunta-3
#Cree una función lambda con la misma funcionalidad que la función de suma que 
# acaba de crear.

suma=lambda a,b,c:a+b+c
suma(4,5,6)
```

```python
#pregunta-4
#-Utilizando la siguiente lista y variable,
#determine si el valor de la variable coincide o no con un valor de la lista...

nombre = 'Enrique'
encontrado = False
lista_nombre = ['Jessica', 'Paul', 'George', 'Henry', 'Adán']
for nombre_lista in lista_nombre:
    if nombre_lista== nombre:
        encontrado=True
        print("✓encontrado")
        break
if not encontrado:
    print("✗no encontrado")
   
```
