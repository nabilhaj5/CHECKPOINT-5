```python
cuadrados = []
for x in range(5):
    cuadrados.append(x**2)
print("Cuadrados LARGO:", cuadrados)
```

```python
cuadrados = [x**2 for x in range(5)]
print("Cuadrados CORTO:", cuadrados)
```

```python
pares = []
for x in range(10):
    if x % 2 == 0:
        pares.append(x)
print("Pares LARGO:", pares)
```

```python
pares = [x for x in range(10) if x % 2 == 0]
print("Pares CORTO:", pares)
```

```python
mayus = []
for l in "python":
    mayus.append(l.upper())
print("Mayúsculas LARGO:", mayus)
```

```python
mayus = [l.upper() for l in "python"]
print("Mayúsculas CORTO:", mayus)
```

```python
grandes = []
for x in [1,5,3,8,2]:
    if x > 5:
        grandes.append(x)
print("Grandes LARGO:", grandes)
```

```python
grandes = [x for x in [1,5,3,8,2] if x > 5]
print("Grandes CORTO:", grandes)
```
